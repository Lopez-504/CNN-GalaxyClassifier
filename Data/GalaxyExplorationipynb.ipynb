```python
import matplotlib.pyplot as plt
import numpy as np

# Load the 'ans' dataset for labels
with h5py.File('/content/galaxies_project3.h5', 'r') as f_read:
    labels = f_read['ans'][()]

# Extract the first 4 images and labels
num_images_to_plot = 4
first_4_images = dataset_data[:num_images_to_plot]
first_4_labels = labels[:num_images_to_plot]

# Plot the images
plt.figure(figsize=(12, 12))
for i in range(num_images_to_plot):
    plt.subplot(1, num_images_to_plot, i + 1)
    plt.imshow(first_4_images[i])
    plt.title(f"Label: {first_4_labels[i]}")
    plt.axis('off')
plt.tight_layout()
plt.show()
```
